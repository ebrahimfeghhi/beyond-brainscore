```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Function for rounding to 2 significant digits
def round_sig(x, sig=2):
    return float(f"{x:.{sig}g}")
```

```python
oasm_omegas = pd.read_csv('figures_data/figure2/oasm_omega_values.csv', index_col=False)
```

```python
color_palette = sns.color_palette(["gray", "blue", 'black'])
for dataset in ['pereira', 'fedorenko', 'blank']:
    fig, ax = plt.subplots(1,1,figsize=(3,4))
    oasm_omegas_dataset = oasm_omegas.loc[oasm_omegas.dataset==dataset]
    for fe in oasm_omegas_dataset.feature_extraction.unique():
        oasm_omegas_dataset_fe = oasm_omegas_dataset.loc[oasm_omegas_dataset.feature_extraction == fe]
        mean_value = oasm_omegas_dataset_fe['values'].mean()
        std_error = oasm_omegas_dataset_fe['values'].std() / np.sqrt(len(oasm_omegas_dataset_fe))

        print(fe, dataset, round_sig(mean_value), round_sig(std_error))
    sns.barplot(oasm_omegas_dataset, hue='feature_extraction', y='values', x='dataset', alpha=0.4, legend=False, errorbar=None, palette=color_palette,ax=ax)
    sns.stripplot(oasm_omegas_dataset, hue='feature_extraction', y='values', x='dataset', dodge=True, size=10, alpha=0.8, legend=False, palette=color_palette,ax=ax)
    sns.despine()
    ax.set_xlabel('')
    ax.set_xticklabels('')
    ax.set_xticks([])
    ax.set_yticks([0,100])
    ax.set_yticklabels([0,100], fontsize=20)
    ax.set_ylabel('')
    fig.savefig(f'/home3/ebrahim2/beyond-brainscore/analyze_results/figures_code/figures/new_figures/figure2/banded/oasm_{dataset}.png', bbox_inches='tight')
    fig.savefig(f'/home3/ebrahim2/beyond-brainscore/analyze_results/figures_code/figures/new_figures/figure2/banded/oasm_{dataset}.pdf', bbox_inches='tight')
    plt.show()
```
