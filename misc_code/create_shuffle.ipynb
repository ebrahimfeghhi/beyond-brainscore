```python
def order_by_passage_length(arr):
    
    # Create a dictionary to group elements by their value
    grouped_elements = {}
    for index, element in enumerate(arr):
        if element not in grouped_elements:
            grouped_elements[element] = []
        grouped_elements[element].append(index)
    
    length_4_passages = []
    length_3_passages = []
    for key, item in grouped_elements.items():
        if len(item) == 4:
            length_4_passages.append(key)
        elif len(item) == 3:
            length_3_passages.append(key)
        else:
            print("Wrong length passage")
            
    order_by_length_indices = []
    for value in length_3_passages:
        order_by_length_indices.extend(grouped_elements[value])
        
    for value in length_4_passages:
        order_by_length_indices.extend(grouped_elements[value])
    
    return order_by_length_indices, length_3_passages, length_4_passages
```

```python
obl, l3, l4 = order_by_passage_length(data_labels)
data_labels_obl = data_labels[obl]
```

```python
def shuffle_l3_l4(arr, num_sets, length_3_passages, length_4_passages):
    
    # Create a dictionary to group elements by their value
    grouped_elements = {}
    for index, element in enumerate(arr):
        if element not in grouped_elements:
            grouped_elements[element] = []
        grouped_elements[element].append(index)
    
    shuffle_indices = []
    
    for _ in range(num_sets):

        # Shuffle the group values to determine their new order
        random.shuffle(length_4_passages)
        random.shuffle(length_3_passages)

        # Create the shuffled indices based on the new order of group values
        shuffled_indices = []
        for value in length_3_passages:
            shuffled_indices.extend(grouped_elements[value])
        for value in length_4_passages:
            shuffled_indices.extend(grouped_elements[value])
        
        
        shuffle_indices.append(shuffled_indices)
    
    return np.vstack(shuffle_indices)
```
